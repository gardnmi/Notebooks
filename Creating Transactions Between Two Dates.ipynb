```python
def create_transactions_between_dates(df, start_date_field_name, end_date_field_name, numeric_field_names ,monthly=False):
    
    df = df.set_index(start_date_field_name, drop=False)
    df = df.reindex(pd.date_range(df.index[0], end=example[end_date_field_name][0]))
    df = df.reindex(pd.date_range(df.index[0], end=df['End Date'][0]))
    df = df.ffill()

    daily_fields = []
    for field in numeric_field_names:
        df[f'Daily {field}'] = df[field] / len(df) 
        daily_fields.append(f'Daily {field}')
    
    if monthly:
        df[[f'Monthly {fieldname}' for fieldname in numeric_field_names]] = df.resample('m')[daily_fields].transform('sum')
        df = df.resample('MS').agg('max')
    
    return df
```

```python
trx_daily = group.apply(create_transactions_between_dates, 'Start Date', 'End Date', ['Price'])
trx_daily.head()
```
