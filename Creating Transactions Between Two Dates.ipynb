```python
def create_transactions_between_dates(group, st_date, end_date, numeric_columns, monthly=False):
    """
    Creates a Transactions between Two Dates.
    
    This function should be passed into the pandas.core.groupby.GroupBy.apply as a paramter.
    
    Example: group.apply(create_transactions_between_dates, 'Start Date', 'End Date', ['Price'])

    Parameters
    ----------
    group : DataFrameGroupBy
        A groupby object created from calling the method .groupby on a dataframe
    
    st_date : str, 
        column name that contains the start date
        
    end_date : str, 
        column name that contains the end date
        
    numeric_column : list, 
        list of columns that you want to aggregate        
        
    monthly : bool, 
        True returns monthly transactions   

    Returns
    -------
    pandas.DataFrame

    """
     
    df = group.set_index(st_date, drop=False)
    df = df.reindex(pd.date_range(df.index[0], end=df[end_date][0]))
    df = df.reindex(pd.date_range(df.index[0], end=df['End Date'][0]))
    df = df.ffill()

    daily_coumns = []
    for column in numeric_columns:
        df[f'Daily {column}'] = df[column] / len(df) 
        daily_coumns.append(f'Daily {column}')
    
    if monthly:
        df[[f'Monthly {column}' for column in numeric_columns]] = df.resample('m')[daily_coumns].transform('sum')
        df = df.resample('MS').agg('max')
    
    return df
```

```python
trx_daily = group.apply(create_transactions_between_dates, 'Start Date', 'End Date', ['Price'])
trx_daily.head()
```
